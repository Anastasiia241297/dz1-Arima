```python
# 1. Импорт библиотек

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import StandardScaler

# 2. Загрузка данных

def load_data(path):
    df = pd.read_csv(path, parse_dates=['Date'])
    df = df.groupby('Date')['number_sold'].sum().reset_index()
    df.set_index('Date', inplace=True)
    return df['number_sold']

train = load_data('train.csv')
test = load_data('test.csv')


# 3. Визуализация исходных данных

plt.figure(figsize=(14, 6))
train.plot(label='Train (2010-2018)', alpha=0.7)
test.plot(label='Test (2019)', color='green', alpha=0.7)
plt.title('Исходные данные о продажах', fontsize=14)
plt.ylabel('Количество продаж')
plt.xlabel('Дата')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

# 4. Создание фич для сезонности

def create_features(df):
    features = pd.DataFrame(index=df.index)
    day_of_year = df.index.dayofyear

    day_norm = (day_of_year - 1) / 365

    features['sin1'] = np.sin(2 * np.pi * day_norm)
    features['cos1'] = np.cos(2 * np.pi * day_norm)
    features['sin2'] = np.sin(4 * np.pi * day_norm)
    features['cos2'] = np.cos(4 * np.pi * day_norm)

    features['day_of_week'] = df.index.dayofweek
    features['quad_trend'] = (day_norm - 0.5)**2

    return features

train_features = create_features(train)
test_features = create_features(test)

# 5. Масштабирование фич

scaler = StandardScaler()
train_features_scaled = scaler.fit_transform(train_features)
test_features_scaled = scaler.transform(test_features)

# 6. Построение и обучение модели

model = SARIMAX(train,
                exog=train_features_scaled,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 7),
                enforce_stationarity=False)

results = model.fit(maxiter=200, disp=True)
print(results.summary())

# 7. Прогнозирование

forecast = results.get_forecast(steps=len(test),
                              exog=test_features_scaled)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# 8. Визуализация прогноза

plt.figure(figsize=(14, 7))
train.plot(label='Historical', alpha=0.7)
test.plot(label='Actual', color='green', alpha=0.7)
forecast_mean.plot(label='Forecast', color='red', linestyle='--')

plt.axhline(y=55300, color='blue', linestyle=':', label='Январь/Дек')
plt.axhline(y=54850, color='orange', linestyle=':', label='Май')
plt.axhline(y=55750, color='purple', linestyle=':', label='Ноябрь')

plt.title('Прогноз продаж с учетом сезонности', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# 9. Анализ остатков

results.plot_diagnostics(figsize=(12, 8))
plt.tight_layout()
plt.show()

# 10. Дополнительная визуализация результатов

# Создаем DataFrame для анализа
results_df = pd.DataFrame({
    'Actual': test,
    'Forecast': forecast_mean,
    'Error': test - forecast_mean
}, index=test.index)

# Добавляем месяц и день недели
results_df['Month'] = results_df.index.month_name()
results_df['Weekday'] = results_df.index.day_name()

# График распределения ошибок
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(results_df['Error'], bins=30, color='skyblue', edgecolor='black')
plt.title('Распределение ошибок прогноза')
plt.xlabel('Ошибка')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.scatter(results_df['Forecast'], results_df['Error'], alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Ошибки vs Прогноз')
plt.xlabel('Прогноз')
plt.ylabel('Ошибка')

plt.tight_layout()
plt.show()

# 11. Анализ по месяцам

monthly_stats = results_df.groupby('Month').agg({
    'Actual': 'mean',
    'Forecast': 'mean',
    'Error': ['mean', 'std']
})

print("\nСтатистика по месяцам:")
print(monthly_stats)

# 12. Оценка качества

def evaluate(actual, predicted):
    error = actual - predicted
    mae = np.mean(np.abs(error))
    rmse = np.sqrt(np.mean(error**2))
    mape = np.mean(np.abs(error/actual))*100

    print(f"MAE: {mae:.1f}")
    print(f"RMSE: {rmse:.1f}")
    print(f"MAPE: {mape:.2f}%")
    print(f"Средний прогноз: {np.mean(predicted):.1f}")
    print(f"Средний факт: {np.mean(actual):.1f}")

print("\nОценка качества:")
evaluate(test, forecast_mean)
```
